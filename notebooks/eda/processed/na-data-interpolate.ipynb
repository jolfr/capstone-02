```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filename = '../../../data/processed/na-data-int.csv'
df = pd.read_csv(filename)
```

```python
def generateScatter(dataframe, variable):
    grouped = dataframe.groupby('zipcode')
    plt.yscale('log')
    for name, group in grouped:
        plot = plt.plot(group['year'], group[variable], linewidth=0.25, alpha=0.1, color='blue')
    return plot
```

```python
generateScatter(df, 'EQI_zip')
_ = plt.title('year vs. EQI_zip by zipcode')
_ = plt.xlabel('year')
_ = plt.ylabel('EQI_zip')
```
